```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = sns.load_dataset("exercise")
print(data)
sns.set_style(style = "darkgrid")
sns.barplot(x = "time", y = "pulse", data =data )
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = sns.load_dataset("exercise")
print(data)
sns.set_style(style = "whitegrid")
sns.barplot(x = "time", y = "pulse", data =data )
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = sns.load_dataset("tips")
a = sns.FacetGrid(data, col = "day")
plt.show()
```

```python
# Get the unique values for the 'day' column to ensure proper order
days_order = data['day'].unique()


a = sns.FacetGrid(data, col="smoker", height=2, hue ="sex")
a.map(sns.barplot, "day", "tip", order=days_order)
plt.show()
```

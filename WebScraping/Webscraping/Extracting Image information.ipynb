```python
import requests
from bs4 import BeautifulSoup
web = requests.get("https://www.shutterstock.com/images")
web
```

```python
soup = BeautifulSoup(web.content, "html.parser")
print(soup.prettify())
```

```python
img = soup.find_all("img")
img
```

```python
for i in img:
    print(i.get("src"))
```

```python
# Import necessary libraries
import numpy as np
importplt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
 pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'students_data.csv'
df = pd.read_csv(file_path)

# Check the first few rows to ensure it's loaded properly
print(df.head())

# Display the column names to verify
print(df.columns)

# Histogram for MathScore
sns.histplot(df['MathScore'], kde=True, color='green')
```

```python
# Convert WklyStudyHours to numeric categories (for better grouping)
df['WklyStudyHours'] = df['WklyStudyHours'].replace({'< 5': 0, '5 - 10': 1, '10 - 15': 2, '> 15': 3})

# Set up the figure
plt.figure(figsize=(5, 3))

# Create a bar plot for average math scores by weekly study hours
sns.barplot(x='WklyStudyHours', y='MathScore', data=df, palette='Blues')

# Add title and labels
plt.title('Average Math Score by Weekly Study Hours', fontsize=16)
plt.xlabel('Weekly Study Hours')
plt.ylabel('Average Math Score')

# Show the plot
plt.show()
```

```python
# Set up the figure
plt.figure(figsize=(10, 6))

# Create a scatter plot of WklyStudyHours vs MathScore
sns.scatterplot(x='WklyStudyHours', y='MathScore', data=df, color='purple')

# Add title and labels
plt.title('Weekly Study Hours vs Math Score', fontsize=16)
plt.xlabel('Weekly Study Hours')
plt.ylabel('Math Score')

# Show the plot
plt.show()
```

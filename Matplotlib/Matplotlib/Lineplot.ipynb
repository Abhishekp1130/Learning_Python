```python
import matplotlib.pyplot as plt
x =["day1", "day2", "day3", "day4", "day5"]
y = [300, 420, 250, 230, 400]
y1 = [500, 450, 300, 250, 320]
plt.plot(x, y, marker="o", ls="--", color="black", label="week1")
plt.plot(x, y1, marker="*", ls="-", color="blue", label="week2", alpha= 0.5)
plt.legend()
plt.show()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
data = pd.read_csv("sample_sales_data.csv")
df = pd.DataFrame(data)
print(df)
grouped_by = df.groupby("Region")["Sales"].sum()
print(grouped_by)

plt.plot(grouped_by.index, grouped_by.values,marker="*", ls="-", color="black")
plt.show()
```

```python
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5, 6, 7]

NOP1 = [5, 10, 30, 20, 35, 60, 80]
NOP2 = [10, 20, 30, 25, 40, 50, 90]
NOP3 = [8, 30, 50, 60, 70, 90, 100]

plt.stackplot(days, NOP1, NOP2, NOP3, colors = ["black", "brown", "blue"], labels =["week1", "week2", "week3"])
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Category': ['Fruits', 'Vegetables', 'Grains', 'Dairy', 'Meat', 'Legumes', 'Nuts', 'Sweets', 'Beverages'],
    'Calories': [52, 41, 370, 150, 250, 200, 607, 450, 80],
    'Protein (g)': [0.3, 1.9, 13, 8, 20, 14, 20, 4, 0.5],
    'Fat (g)': [0.2, 0.2, 3.5, 8, 15, 1, 54, 22, 0],
    'Carbs (g)': [14, 9, 75, 12, 0, 35, 13, 60, 20],
    'Fiber (g)': [2.4, 3.6, 12, 0, 0, 15, 7, 1, 0]
}

df = pd.DataFrame(data)

grouped = df.groupby('Category')[['Calories', 'Protein (g)', 'Fat (g)']].mean()
plt.stackplot(grouped.index, grouped['Calories'], grouped['Protein (g)'], grouped['Fat (g)'])
plt.show()
```

```python
data = {
    'Category': ['Fruits', 'Vegetables', 'Grains', 'Dairy', 'Meat', 'Legumes', 'Nuts', 'Sweets', 'Beverages'],
    'Calories': [52, 41, 370, 150, 250, 200, 607, 450, 80],
    'Protein (g)': [0.3, 1.9, 13, 8, 20, 14, 20, 4, 0.5],
    'Fat (g)': [0.2, 0.2, 3.5, 8, 15, 1, 54, 22, 0],
    'Carbs (g)': [14, 9, 75, 12, 0, 35, 13, 60, 20],
    'Fiber (g)': [2.4, 3.6, 12, 0, 0, 15, 7, 1, 0]
}

df = pd.DataFrame(data)


grouped = df.groupby('Category')[['Calories', 'Protein (g)', 'Fat (g)']].mean()


plt.figure(figsize=(10, 6))
plt.stackplot(grouped.index, grouped['Calories'], grouped['Protein (g)'], grouped['Fat (g)'], 
              labels=['Calories', 'Protein (g)', 'Fat (g)'])
plt.legend(loc='upper left')
plt.title('Mean Nutritional Content by Food Category')
plt.xlabel('Food Category')
plt.ylabel('Nutritional Values (Mean)')
plt.xticks(rotation=45)
plt.show()
```

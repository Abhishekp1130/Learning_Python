```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [45, 34, 56, 23, 45]
plt.subplot(2, 2, 1) #rows #columns #chartnumber
plt.plot(x,y)



x = [5, 6, 7, 8, 9]
y = [67, 50, 66, 56, 82]
plt.subplot(2, 2, 2)
plt.bar(x, y)



x = [2, 4, 6, 8, 10]
y = [57, 60, 67, 76, 92]
plt.subplot(2, 2, 3)
plt.scatter(x, y)


x = [2, 4, 6, 8, 10]
y = [67, 50, 47, 77, 82]
plt.subplot(2, 2, 4)
plt.step(x, y)
plt.show()
```

```python
x = [2, 4, 6, 8, 10]
y = [57, 60, 67, 76, 92]
plt.subplot(1, 2, 1)
plt.title("Earning")
plt.bar(x, y)


x = [2, 4, 6, 8, 10]
y = [67, 50, 47, 77, 82]
plt.subplot(1, 2, 2)
plt.title("Spending")
plt.plot(x, y)
plt.suptitle("employee data")
plt.show()
```

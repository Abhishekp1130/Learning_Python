```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [45, 67, 33, 67, 12]
y1 = [41, 60, 13, 66, 13]
y2 = [11, 30, 55, 60, 19]

plt.figure(figsize = [8, 5])
plt.plot(x, y, label="male")
plt.plot(x, y1, label="female")
plt.plot(x, y2, label="kids")
plt.legend(["M", "F","K"])
plt.show()
```

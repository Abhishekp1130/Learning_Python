```python
import matplotlib.pyplot as plt
y = [98, 67, 88, 95, 88]
x = ["Part1", "Part2", "Part3", "Part4", "Part5"]

plt.bar(x , y, color="black")
plt.xlabel("Parts Of Harry Potter", fontsize = 20)
plt.ylabel("Popularity", fontsize = 20)
plt.title("Popularity Of Different Parts Of Harry Potter", fontsize = 20)
plt.show()
```

```python
import matplotlib.pyplot as plt
y = [98, 67, 88, 95, 88]
x = ["Part1", "Part2", "Part3", "Part4", "Part5"]
colors = ["red", "blue", "orange", "black", "green"]
plt.bar(x , y, color=colors, edgecolor="black")
plt.xlabel("Parts Of Harry Potter", fontsize = 20)
plt.ylabel("Popularity", fontsize = 20)
plt.title("Popularity Of Different Parts Of Harry Potter", fontsize = 20)
plt.savefig("Bar.png")
plt.show()
```

```python
# to load the file and perform
import pandas as pd 
import matplotlib.pyplot as plt
data = pd.read_csv("sample_sales_data.csv")
df = pd.DataFrame(data)
print(df)
grouped_by = df.groupby("Region")["Sales"].sum()
print(grouped_by)
plt.bar(grouped_by.index, grouped_by.values)
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.title("Total Sales by Region")

plt.show()

```

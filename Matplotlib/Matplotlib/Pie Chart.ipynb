```python
import matplotlib.pyplot as plt
brands = ["Oneplus", "Apple", "Samsung", "Nokia", "Redmi"]
x = [22, 35, 30, 3, 10]
c=["red", "silver", "blue", "pink", "orange"]
ex = [0, 0,0,0,0.1]
plt.pie(x, labels = brands, colors = c, explode=ex, shadow = True, autopct = "%.f", startangle=90)
plt.show()
```
